```python
import pandas as pd
from sklearn import svm
from sklearn import preprocessing
```

```python
df = pd.read_csv('data.csv')
```

```python
status_df = pd.get_dummies(df.status)
```

```python
sum(df.location.isnull())
```

```python
pd.get_dummies(df.location)
```

```python
df = pd.concat([df,status_df],axis='columns')
```

```python
df = df.drop(['status'],axis='columns')
```

```python
df
```

```python
regr = svm.SVR(kernel = 'linear', C = 1.0)
regr.fit(df[['area','bhk','Ready to move','Under Construction']], df.price)
```

```python
regr.score(df[['area','bhk','Ready to move','Under Construction']], df.price)
```

```python
regr2 = svm.SVR(kernel = 'linear', C = 1.0)
regr2.fit(df[['area','bhk','Ready to move']], df.price)
regr2.score(df[['area','bhk','Ready to move']], df.price)
```
