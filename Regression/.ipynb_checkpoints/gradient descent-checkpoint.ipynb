```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([3,6,9,12,15])
```

```python
m = b = 0
```

```python
iterations = 1000
n = len(x)
lr = 0.001
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

import celluloid
from celluloid import Camera
fig = plt.figure(figsize=(10,10)) # create figure
labelsize_ = 14
camera = Camera(fig)  # create camera
```

```python
for i in range (iterations):
    yPredict = m*x + b
    cost = (1/n)*sum([v**2 for v in (y-yPredict)])
    md = (-2/n)*sum(x*(y-yPredict))
    bd = (-2/n)*sum(y-yPredict)
    m = m - lr*md
    b = b - lr*bd
    print("interation "+ str(i)+" m,b ==>" +str(m) +"/"+str(b) + " cost ==> " + str(cost))

    
    ax1=fig.add_subplot(3, 2, 2) 
    ax1.plot(x,yPredict, color='blue', linestyle="dashed", alpha=0.5)
    ax1.set_title("w", fontsize=17)
    ax1.tick_params(axis='both', which='major', labelsize=labelsize_)
    
    plt.tight_layout()
    
    camera.snap() # take snapshot after each frame/iteration
    
animation = camera.animate(interval = 5,repeat = False, repeat_delay = 500) # create animation 
animation.save('SimpleLinReg_1.gif', writer = 'imagemagick') # save animation
    
    
```
